```python
from raptor import Raptor, Analysis
import numpy as np
import cv2 as cv
import matplotlib
```

```python
[i for i in dir(cv) if i.startswith('COLOR_RGB2')]
```

```python
a_rgb = Analysis('28-12-2021_0_rgb.data', path="data/")
```

```python
a_rgb.data['data'][0].keys()
```

```python
bgr_array = np.array(a_rgb.data['data'][0]['image'])
matplotlib.pyplot.imshow(bgr_array)
bgr_array[:1]
```

```python
bgr_array = bgr_array.astype('float32')
bgr_array = bgr_array / 255
matplotlib.pyplot.imshow(bgr_array)
bgr_array[:1]
```

```python
rgb_array = cv.cvtColor(rgb_array, cv.COLOR_BGR2RGB)
matplotlib.pyplot.imshow(rgb_array)
```

```python
hsv_array = cv.cvtColor(rgb_array, cv.COLOR_RGB2HSV)
matplotlib.pyplot.imshow(hsv_array)
hsv_array[:10]
```

```python
yellow = np.uint8([[[0,255,255 ]]])
hsv_yellow = cv.cvtColor(yellow,cv.COLOR_BGR2HSV)
print( hsv_yellow )
```

```python
hsv_array[420][300]
```

```python
lower_yellow = np.array([10, 25/255, 25/255])
upper_yellow = np.array([200, 255/255, 255/255])

mask = cv.inRange(hsv_array, lower_yellow, upper_yellow)
mask = ~mask
matplotlib.pyplot.imshow(mask)
res = cv.bitwise_and(rgb_array, rgb_array, mask=mask)
```
